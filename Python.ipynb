## Q5) (5 Points)

#### Create a function named as "case_sensitive_sorting" that should:-
- *Accept **only one word (string)** as an input.*
- *Sort that word such that upper cases are sorted only where upper cases were present in the original word.*
- *Similarly, smaller cases are sorted only where the smaller caseswere present in the original word.*
- *Return that case sensitve sorted word*

**Example:-**        
**Input:** *TnuKi*      
**Output:** *KinTu*

````python
# Create Function
def case_sensitive_sorting(word):
    lower_case = ''.join(sorted([c for c in word if c.islower()]))  # Sorts only the lower case letters
    upper_case = ''.join(sorted([c for c in word if c.isupper()]))  # Sorts only the upper case letters

    sorted_word = ''
    lower_index, upper_index = 0, 0

    for char in word:
        if char.islower():
            sorted_word += lower_case[lower_index]
            lower_index += 1
        else:
            sorted_word += upper_case[upper_index]
            upper_index += 1

    return sorted_word

# Test the function
user_input = input("Enter a word: ")
result = case_sensitive_sorting(user_input)
print("Output:", result)


````

#### Create a binomial expansion calculator such that if user inputs the values of a, b, n for (ax + b)^n       
#### where n is a positive integer then your calculator should output its expansion.

#### *For example :-*
```
Input:- (1 - x)^2
Output:- x^2 - 2x + 1
```
#### **NOTE:-**       
#### **1)** *You are not allowed to use any python libraries/modules for this question*      
#### **2)** *Use Pascal's triangle to get the binomial coefficients of each term in the expansion*     
#### **3)** *Using any formula for calculating binomial coefficients is strictly prohibited!*

#### **Below is an image to give you an idea about how Pascal's triangle relates with Binomial expansion:-**
<img src="https://drive.google.com/uc?id=1YnMgziwCHCLk2mc-msMdeLc99FevtIWr" height = 300 width = 100%>

````python
# Create a function that would get you binomial coefficients for any (ax + b)^n
def binomial_expansion(a, b, n):
    # Initialize an empty list to store the expansion terms
    expansion_terms = []

    # Calculate the binomial coefficients using Pascal's triangle
    for k in range(n + 1):
        coefficient = 1
        for i in range(k):
            coefficient *= (n - i)
            coefficient //= (i + 1)  # Integer division to avoid floating-point results
        # Append the term to the list
        term = coefficient * (a ** k) * (b ** (n - k))
        expansion_terms.append(term)

    # Construct the final expression
    expression = ""
    for k, term in enumerate(expansion_terms):
        if k > 0:
            expression += " + " if term >= 0 else " - "
        expression += f"{abs(term)}x^{k}"

    return expression

# Take user input for expression
a = float(input("Enter the value of a: "))
b = float(input("Enter the value of b: "))
n = int(input("Enter a positive integer value for n: "))

result = binomial_expansion(a, b, n)
print(f"Input: ({a}x + {b})^{n}")
print(f"Output: {result}")


````

## Q11 (5 points)
#### Complete and correct the below given function to generate a sin wave for 5 seconds duration

````python
import numpy as np
from matplotlib import pyplot as plt

SAMPLE_RATE = 44100  # Hertz
DURATION = 5  # Seconds

def generate_sine_wave(freq, sample_rate, duration):
    # Generate a sequence of time values for the duration at the given sample rate
    x = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    frequencies = x * freq
    # Use sine function to generate the sine wave
    y = np.sin(2 * np.pi * frequencies)
    return x, y

# Generate a 2 hertz sine wave that lasts for 5 seconds
x, y = generate_sine_wave(2, SAMPLE_RATE, DURATION)
plt.plot(x, y)
plt.show()
````

## Q15 (5 points)
Given below is an example of "fits" data file format. Explain about each column - Name, Ver, Type, Cards, Dimensions and Format.


```
No.    Name      Ver    Type      Cards   Dimensions   Format
  0  PRIMARY       1 PrimaryHDU     467   ()      
  1  SCI           1 ImageHDU       215   (690, 690)   float32   
  2  WHT           1 ImageHDU       215   (690, 690)   float32   
  3  CTX           1 ImageHDU       215   (690, 690)   int32
```

Ans
1. No. - This column represents the index or number assigned to each extension in the FITS file.
2. Name-The name of the extension. eg: PRIMARY: The primary HDU (Header Data Unit) is the main header that contains global information about the entire FITS file.
3. Ver: The version number of the extension. In this eg., all extensions have version 1.
4. Type: The type of HDU
5. Cards: The number of header cards or header keywords in the extension header.
6. Dimensions: The dimensions of the data array associated with the extension Eg: (690,690)=2D image data ,()=no imade data
7. Format: The data format of the array elements in the extension. Common formats include:
Eg: float32: 32-bit floating-point ,int32: 32-bit integers.
